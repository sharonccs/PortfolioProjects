#1 Create a new 'Type of Cereal' column in your dataframe (1 point) by copying the 'name' column. Write a function to replace the names of the cereal in your new column with one of these categories Bran, Wheat, Fiber, Protein, Crunch, Corn, Nut, Rice and Other (3 points). Hint: the function should look through the text in the cereal name and determine, based on its contents, how to categorize the cereal type.

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'C:\Users\User\Desktop\cereal.csv')
df
```

```python
def convention(x):
    if 'Bran' in x['name']:
        return 'Bran'
    elif 'Wheat' in x['name']:
        return 'Wheat'
    elif 'Fiber' in x['name']:
        return 'Fiber'
    elif 'Protein' in x['name']:
        return 'Protein'
    elif 'Crunch' in x['name']:
        return 'Crunch'
    elif 'Corn' in x['name']:
        return 'Corn'
    elif 'Nut' in x['name']:
        return 'Nut'
    elif 'Rice' in x['name']:
        return 'Rice'
    else:
        return 'Other'
```

```python
df['Type_of_Cereal'] = df.apply(convention, axis=1)
```

```python
with pd.option_context("display.max_rows", None, "display.max_columns", None):
  display(df)
```

#2 Identify the negative values in the data set and replace them with the median value for that column.

```python
df.median()
```

```python
df['carbo'].median()
```

```python
def calculation(s):
    if (s['carbo'] < 0) :
        return df['carbo'].median()
    else:
        return s['carbo'] 
df ['carbo']= df.apply(calculation,axis=1)
```

```python
def calculation(s):
    if (s['sugars'] < 0) :
        return df['sugars'].median()
    else:
        return s['sugars'] 
df ['sugars']= df.apply(calculation,axis=1)
```

```python
def calculation(s):
    if (s['potass'] < 0) :
        return df['potass'].median()
    else:
        return s['potass'] 
df ['potass']= df.apply(calculation,axis=1)
```

```python
with pd.option_context("display.max_rows", None, "display.max_columns", None):
  display(df)
```

#3 Standardize the 'weight' column to 1. For this question, you will need to write a function to divide the remaining columns which contain nutritional information by the corresponding value in the weight column, and you will need to divide the value in the weight column by itself in order to get 1. For example, if an observation has a weight value of 1.33 and a calories value of 250, if you divide 250/1.33 you should get a calories value of 188 and a weight value of 1.

```python
def calculation(s):
    if (s['weight'] != 1) :
        return 1
    else:
        return 1
df ['weight']= df.apply(calculation,axis=1)
```

```python
with pd.option_context("display.max_rows", None, "display.max_columns", None):
  display(df)
```

#4 Create a new column to categorize cereals as 'healthy' vs. 'unhealthy'. You can define your own version of healthy vs. unhealthy, or you can use the following: healthy cereals can be defined as those which have low calories (<100), low sodium (<150), low sugar (<9) high fiber (>3), and high protein (>2). All other cereals are unhealthy.

```python
def conditions(s):
    if (s['calories'] < 100) and (s['sodium'] <150) and (s['sugars'] <9 ) and  (s['fiber'] <3 )  :
        return 'healthy'
    else:
        return 'unhealthy'
```

```python
df['Class'] = df.apply(conditions, axis=1)
```

```python
with pd.option_context("display.max_rows", None, "display.max_columns", None):
  display(df)
```
